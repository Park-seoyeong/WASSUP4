## quiz: 네이버 증권 주요뉴스의 제목과 url 모두를 추출하여라

````python
# 슬라이싱 필요
taglist['href']
````

````python
for i in taglist:
    url = PATH[:-1] + i['href']
    print(url)
````

````python
for i in taglist:
    title =  i.text
    print(title)
````

````python
news_title = []
news_url = []

for i in taglist:
    title =  i.text
    url = urljoin(PATH, i['href'])
    print(title, url)
    news_title.append(title)
    news_url.append(url)
````

````python
import pandas as pd
````

````python
df = pd.DataFrame( {'제목': news_title, '주소':news_url } )
df
````
